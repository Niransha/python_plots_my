```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset "AAA.dat" as a DataFrame
df = pd.read_csv('/mnt/rna/home/nkumarachchi2019/dev_1d_ABGEZ_benckmark_systems/beta_newmodel_beta_systems/2KOC_orgi/torsion_probabilities/AllTorsions1.dat', header='infer', delim_whitespace=True, error_bad_lines=False)

# Set the Seaborn theme
sns.set_theme(style="whitegrid")
dt = 0.00002
# Plot the responses for different columns

# Create a figure for the small plot
plt.figure(figsize=(50, 8), dpi=100)

sns.lineplot(x=df.iloc[:, 0] * dt, y=df.iloc[:, 1], label="alpha:6", data=df)
sns.lineplot(x=df.iloc[:, 0] * dt, y=df.iloc[:, 2], label="beta:6", data=df)
sns.lineplot(x=df.iloc[:, 0] * dt, y=df.iloc[:, 3], label="gamma:6", data=df)

# Calculate and plot the Simple Moving Average (SMA) for column 1
sma_window = 200  

df['SMA_alpha:6'] = df.iloc[:, 1].rolling(window=sma_window).mean()
sns.lineplot(x=df.iloc[:, 0] * dt, y=df['SMA_alpha:6'], label=f"SMA_alpha:6 (Window {sma_window})", data=df)

# Calculate and plot the Simple Moving Average (SMA) for column 2
df['SMA_beta:6'] = df.iloc[:, 2].rolling(window=sma_window).mean()
sns.lineplot(x=df.iloc[:, 0] * dt, y=df['SMA_beta:6'], label=f"SMA_beta:6 (Window {sma_window})", data=df)

# Calculate and plot the Simple Moving Average (SMA) for column 3
df['SMA_gamma:6'] = df.iloc[:, 3].rolling(window=sma_window).mean()
sns.lineplot(x=df.iloc[:, 0] * dt, y=df['SMA_gamma:6'], label=f"SMA_gamma:6 (Window {sma_window})", data=df)

plt.legend()
plt.show()
```
