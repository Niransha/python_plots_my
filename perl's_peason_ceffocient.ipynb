```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is in a DataFrame named 'df'
# You may need to adjust column names based on your actual data

# Sample Data
'''
data = {
    'replicate': ['replicate_1', 'replicate_2', 'replicate_1', 'replicate_2', 'replicate_1', 'replicate_2', 'replicate_1', 'replicate_2', 'replicate_1', 'replicate_2', 'replicate_1', 'replicate_2', ],
    'Hormone': ['Hormone_1', 'Hormone_1', 'Hormone_1', 'Hormone_1', 'Hormone_1','Hormone_1', 'Hormone_0', 'Hormone_0', 'Hormone_0', 'Hormone_0', 'Hormone_0', 'Hormone_0', 'Hormone_0', ],
    'Cutting_age': ['Cutting_age_1', 'Cutting_age_1', 'Cutting_age_2', 'Cutting_age_2', 'Cutting_age_3', 'Cutting_age_3', 'Cutting_age_1', 'Cutting_age_1', 'Cutting_age_2', 'Cutting_age_2', 'Cutting_age_3', 'Cutting_age_3'],
    'response': [0, 0, 30 , 20, 20, 10, 0, 10, 10, 40, 80, 50]
}
'''

data = {
     'Hormone': ['Hormone_1', 'Hormone_1', 'Hormone_1', 'Hormone_1', 'Hormone_1','Hormone_1', 'Hormone_0', 'Hormone_0','Hormone_0','Hormone_0','Hormone_0','Hormone_0',],
    'Cutting_age': ['Cutting_age_1', 'Cutting_age_1', 'Cutting_age_2', 'Cutting_age_2', 'Cutting_age_3', 'Cutting_age_3', 'Cutting_age_1', 'Cutting_age_1', 'Cutting_age_2', 'Cutting_age_2', 'Cutting_age_3', 'Cutting_age_3'],
    'response': [0, 0, 30 , 20, 20, 10, 0, 10, 10, 40, 80, 50]
}



df = pd.DataFrame(data)

# Encode categorical variables as numerical values
df_encoded = pd.get_dummies(df, columns=['Hormone', 'Cutting_age'])

# Calculate correlation matrix
correlation_matrix = df_encoded.corr()


# Extract and rank correlation values (excluding 'response')
correlation_ranking = correlation_matrix['response'].drop('response').abs().sort_values(ascending=False)

#Print the ranking
print("Correlation Ranking:")
print(correlation_ranking)

# Remove the 'response' row and column from the correlation matrix
correlation_matrix = correlation_matrix.drop('response', axis=0)
correlation_matrix = correlation_matrix.drop('response', axis=1)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
